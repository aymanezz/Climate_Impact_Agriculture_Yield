```python
# Group by 'YEAR' and 'SUBDIVISION' to calculate the mean for each subdivision within each year
grouped_subdivisions = df.groupby(['YEAR', 'SUBDIVISION'])['ANNUAL'].mean().reset_index()
grouped_subdivisions
```

```python

# Group by 'YEAR' to calculate the overall mean for each year
average_mean_per_year = grouped_subdivisions.groupby('YEAR')['ANNUAL'].mean().reset_index()

# Rename columns for clarity
average_mean_per_year.columns = ['YEAR', 'AVERAGE_MEAN_ANNUAL']
average_mean_per_year.rename(columns = {'YEAR':'Year'},inplace = True)
```

```python
production = pd.read_csv(r'D:\git\Climate_Impact_Agriculture_Yield\data\modified\clean_productions.csv')
```

```python
temp_change = pd.read_csv(r'D:\git\Climate_Impact_Agriculture_Yield\data\modified\clean_temperature_change.csv')
land_cover = pd.read_csv(r'D:\git\Climate_Impact_Agriculture_Yield\data\modified\clean_land_cover.csv')
sea_level = pd.read_csv(r'D:\git\Climate_Impact_Agriculture_Yield\data\modified\clean_sea_level.csv')
atmospheric = pd.read_csv(r'D:\git\Climate_Impact_Agriculture_Yield\data\modified\clean_Atmospheric.csv')
```

```python
# Merge temperature change data
production = pd.merge(production, temp_change, how='left', on=['Area', 'Year'])

# Merge land cover data
production = pd.merge(production, land_cover, how='left', on=['Area', 'Year'])

# Merge atmospheric data
production = pd.merge(production, atmospheric, how='left', left_on='Year', right_on='Year')

# Merge atmospheric data
production = pd.merge(production, sea_level, how='left', left_on='Year', right_on='Year')
```

```python
produc_india = production[production['Area'] == 'India']
```

```python
produc_india.reset_index(drop = True,inplace = True)
produc_india
```

```python
precip_produc = pd.merge(produc_india,average_mean_per_year,on = 'Year')
```

```python
precip_produc
```

```python
# Creating lag features
for lag in [1, 3, 6]:
    production[f'Temperature Change_lag_{lag}'] = production.groupby('Area')['Temperature Change'].shift(lag)
    production[f'CO2concentration_lag_{lag}'] = production['CO2concentration'].shift(lag)

# Creating interaction features
production['Temperature Change_x_Land Cover'] = production['Temperature Change'] * production['Covered']
production['Temperature Change_x_CO2concentration'] = production['Temperature Change'] * production['CO2concentration']
production['Land Cover_x_CO2concentration'] = production['Covered'] * production['CO2concentration']
```

```python
production.dropna(inplace = True)
```

```python
production.head()
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Assuming `production` is your DataFrame
# Drop columns that you don't need
process_data = production.drop(columns=['Area', 'Item', 'Indicator'])

# Identify categorical and numerical columns
categorical_columns = process_data.select_dtypes(include=['object']).columns
numerical_columns = process_data.select_dtypes(include=['number']).columns

# Define the transformers
categorical_transformer = OneHotEncoder(handle_unknown='ignore')
numerical_transformer = StandardScaler()

# Create a ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_columns),
        ('cat', categorical_transformer, categorical_columns)
    ])

# Fit and transform the data
processed_data = preprocessor.fit_transform(process_data)

# Convert to DataFrame if needed (optional)
# Note: ColumnTransformer returns a NumPy array
processed_data_df = pd.DataFrame(processed_data, columns=preprocessor.get_feature_names_out())

# Display the processed data
processed_data_df.columns
```
