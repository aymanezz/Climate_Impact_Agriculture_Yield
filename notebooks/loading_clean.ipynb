i will focus only on the temperature change no need for standard deviation for now 

```python
temp_df = temp_df[temp_df['Element'] != 'Standard Deviation']
temp_df.drop(columns = ['Element'],inplace = True)
```

```python
temp_df.info()
```

```python
non_numeric_cols = temp_df.select_dtypes(exclude=[np.number]).columns
print("\nColumns with non-numeric values:", non_numeric_cols.tolist())
```

```python
def fill_na_with_row_mean(row):
    # Filter numeric values
    numeric_vals = row[row.apply(lambda x: np.issubdtype(type(x), np.number))]
    # Calculate the mean of the numeric columns in the row
    mean_val = numeric_vals.mean()
    # Fill NaNs in numeric columns with the mean value
    return row.apply(lambda x: mean_val if pd.isna(x) and np.issubdtype(type(x), np.number) else x)
```

```python
temp_df = temp_df.apply(fill_na_with_row_mean,axis = 1)
```

```python
temp_df.info()
```

```python
temp_df.head()
```

```python
temp_df.Months.unique()
```

```python
temp_df = temp_df[temp_df['Months'] == 'Meteorological year']
temp_df.drop(columns = ['Months'],inplace = True)
```

```python
temp_df.head()
```

```python
temp_df = pd.melt(temp_df,id_vars = ['Area'],var_name = 'Year',value_name = 'Temperature Change')
```

```python
temp_df.head()
```

```python
temp_df['Year'] = temp_df['Year'].str.replace('Y','')
```

```python
temp_df.head()
```

```python
temp_df = temp_df.drop_duplicates()
temp_df.to_csv(r'D:\git\Climate_Impact_Agriculture_Yield\data\modified\clean_temperature_change.csv',index = False)
```

i will fill na values with the mean of rows not columns

```python
start_row = 2
end_row = 30
def fill_nan_with_row_mean_specific_range(row):
    row_range = row.iloc[start_row:end_row+1]  # Select the specific range of columns
    row_mean = row_range.mean()  # Calculate row-wise mean for the range
    return row.fillna(row_mean)  # Fill NaN with the row-wise mean

```

```python
df_filled_land_cover = land_cover_df.apply(fill_nan_with_row_mean_specific_range, axis=1)
print("\nDataFrame after filling NaN with row-wise mean for rows 2 to 30:")
df_filled_land_cover
```
